## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI
import pandas as pd

df = pd.read_csv('datasets\movies.csv')

df['US Gross'].max()
df.loc[df['US Gross'] == 760167650.0]
```

### Qual o diretor dirigiu mais filmes?

```python
# ADICIONE O SEU CÓDIGO AQUI

df['Director'].mode()

df.loc[df['Director'] ==  "Steven Spielberg"][["Title", "Director"]]
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI

mais_lucrativos = df['US Gross'] / df['Production Budget']

df['mais_lucrativos'] = mais_lucrativos

df.sort_values(by="mais_lucrativos", ascending=False)[["Title", "mais_lucrativos"]]
```
